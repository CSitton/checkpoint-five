```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np


small_df = pd.read_csv("Lessthan_5_victim_count.csv")
mid_df = pd.read_csv("5_to_14_victim_count.csv")
mod_df = pd.read_csv("15_to_30_victim_count.csv")
large_df = pd.read_csv("Highest_victim_count.csv")
```

## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm 

X= large_df['Country']
y= large_df['Proven victims']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)

X_train
y_train
```

```python
X = pd.get_dummies(X, drop_first=True)

X= sm.add_constant(X)

model = sm.OLS(y,X).fit()

print(model.summary())
print(model.params)
```

## Error Terms

Finally, plot your error terms!

```python
y_train_pred = model.predict(X )

res = (y_train - y_train_pred)

fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```
